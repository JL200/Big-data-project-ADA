## ML Model

```
import pyspark.sql.functions as F
from pyspark.ml.feature import StandardScaler
from pyspark.ml.regression import LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.sql.types import FloatType
```

```
# contagem de linhas
print(df_completo_filtrado.count())
```

```
# valores únicos de uma coluna
print(df_completo_filtrado.select("`Temperature (C)`").distinct().rdd.flatMap(lambda x: x).collect())
```

```
# remover colunas
columns_to_drop = ["Time (s)", "CO (ppm)", "Humidity (%r.h.)", "R4 (MOhm)", "R5 (MOhm)", "R6 (MOhm)", "R7 (MOhm)", "R8 (MOhm)", "R9 (MOhm)", "R10 (MOhm)", "R11 (MOhm)", "R12 (MOhm)", "R13 (MOhm)", "R14 (MOhm)"]
df_completo_filtrado = df_completo_filtrado.drop(*columns_to_drop)
```

```
# converter tipos de dados para float
for col_name in df_completo_filtrado.columns:
    df_completo_filtrado = df_completo_filtrado.withColumn(col_name, df_completo_filtrado[col_name].cast(FloatType()))
```

```
# informações sobre os tipos de dados e colunas nulas
df_completo_filtrado.printSchema()
df_completo_filtrado.select([F.count(F.when(F.isnan(c) | F.col(c).isNull(), c)).alias(c) for c in df_completo_filtrado.columns]).show()
```

```
# separar features e target
X = df_completo_filtrado.drop("Temperature (C)")
y = df_completo_filtrado.select("Temperature (C)")

# X = df_completo_filtrado
# y = df_completo_filtrado
```

```
# dividir dados em treinamento e teste
# X_train, X_test, y_train, y_test = X.randomSplit([0.8, 0.2], seed=42)

X_train, X_test = X.randomSplit([0.8, 0.2], seed=42)
y_train, y_test = y.randomSplit([0.8, 0.2], seed=42)
```

```
# padronização dos dados
# scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures", withStd=True, withMean=False)
# scaler_model = scaler.fit(X_train)
# X_train_std = scaler_model.transform(X_train)
# X_test_std = scaler_model.transform(X_test)

from pyspark.ml.feature import VectorAssembler

assembler = VectorAssembler(inputCols=["Flow Rate (mL/min)", "Temperature (C)", "Heater voltage (V)", "R1 (MOhm)", "R2 (MOhm)", "R3 (MOhm)"], outputCol="features")
X_train = assembler.transform(X_train)
X_test = assembler.transform(X_test)

from pyspark.ml.feature import StandardScaler

scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures", withStd=True, withMean=False)
scaler_model = scaler.fit(X_train)
X_train_std = scaler_model.transform(X_train)
X_test_std = scaler_model.transform(X_test)
```

```
# treinamento do modelo
# lr = LinearRegression(featuresCol="scaledFeatures", labelCol="Temperature (C)", maxIter=10, regParam=0.3, elasticNetParam=0.8)
# model = lr.fit(X_train_std)

# treinamento do modelo
# lr = LinearRegression(featuresCol="scaledFeatures", labelCol="Flow Rate (mL/min)", maxIter=10, regParam=0.3, elasticNetParam=0.8)
# model = lr.fit(X_train_std)

lr = LinearRegression(featuresCol="scaledFeatures", labelCol="Temperature (C)", maxIter=10, regParam=0.3, elasticNetParam=0.8)
model = lr.fit(X_train_std)
```

```
# previsões
y_pred = model.transform(X_test_std)
```

```
# avaliação do modelo
# evaluator = RegressionEvaluator(labelCol="scaledFeatures", predictionCol="Temperature (C)", metricName="r2")
# r2 = evaluator.evaluate(y_pred)
# print("R2 score:", r2)

# avaliação do modelo
evaluator = RegressionEvaluator(labelCol="Temperature (C)", predictionCol="prediction", metricName="r2")
r2 = evaluator.evaluate(y_pred)
print("R2 score:", r2)
```
